```python
import numpy as np
import matplotlib.pyplot as plt
def main(a,b):
    y, x = np.ogrid[-5:5:100j, -5:5:100j]
    plt.contour(x.ravel(), y.ravel(), pow(y, 2) - pow(x, 3) - x * a - b, [0])
    plt.grid()
    plt.show()

if __name__ == '__main__':
    list_of_coords = [
        [1, 1],
        [-25, 0],
        [1, 6],
        [-4, 0],
        [0,-1]
    ]
    for coords in list_of_coords:
        main(coords[0], coords[1])
```

```python
from tinyec.ec import SubGroup, Curve

field = SubGroup(p=17, g=(15, 13), n=18, h=1)
curve = Curve(a=0, b=7, field=field, name='secp256k1')
print('curve:', curve)

for k in range(0, 25):
    p = k * curve.g
    print(f"{k} * G = ({p.x}, {p.y})")
```

```python
def point_doubling(x1, y1, a, p):
    delta = ((3*(x1**2)) +a)/ 2*y1

    x2 = (delta**2 - 2*x1)%p
    y2 = ((delta*(x1 - x2) )-y1)%p

    return x2, y2

def point_addition(x1, y1, x2, y2, p):
    delta = (y2 - y1) / (x2 - x1)
    x3 = ((delta **2) - x1 - x2)%p

    y3 = (delta*(x1-x3) - y1)%p
    
    return x3, y3
```
